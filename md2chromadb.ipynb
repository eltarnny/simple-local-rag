### Insert records into ChromaDB

```python
from tools import *
import chromadb
from customEmbedFunction import TransformerEmbeddingFunction
```

```python
# SET GAME NAME AND FILE HERE

# game_name = 'The Witcher: Old World'
game_name = 'Dune: Imperium - Uprising'

# fname_no_ext = "TWOW_07_Rulebook_Main_(36p)_EN_preview_22-10-20-skompresowany"
fname_no_ext = "DUNE_IMPERIUM_UPRISING_RULEBOOK"

fname = fname_no_ext + '.md'
```

```python
content = read_text_file(fname)
```

Embeddings

For the embedding model,  
from https://huggingface.co/spaces/mteb/leaderboard (STS tab),  
I chose `w601sxs/b1ade-embed`, a small but capable model.  

```python
chunks = split_text(content, max_words=300, overlap=0)

embedder_name = "w601sxs/b1ade-embed"
```

```python
chroma_client = chromadb.PersistentClient(path="chroma")
collection = chroma_client.get_or_create_collection(name='bg-rag', embedding_function=TransformerEmbeddingFunction(embedder_name, use_cuda=True))
```

```python
# This should be a csv or parquet file at some point.
games_metadata = {
    'The Witcher: Old World' :{
        'bg_name': 'The Witcher: Old World',
        'bgg_id': 331106,
        'rules_forum_id': 3544580,
        'general_forum_id': 3544579,
    },
    'Dune: Imperium - Uprising' :{
        'bg_name': 'Dune: Imperium - Uprising',
        'bgg_id': 397598,
        'rules_forum_id': 4410830,
        'general_forum_id': 4410829,
    },
}

# From these values, the only one needed for now is bg_name
bg_name = games_metadata[game_name]['bg_name']
bgg_id = games_metadata[game_name]['bgg_id']
rules_forum_id = games_metadata[game_name]['rules_forum_id']
general_forum_id = games_metadata[game_name]['general_forum_id']
```

```python
# Here, I manually add some metadata for each game, that might be needed in the future.

num_chunks = len(chunks)
print(num_chunks)
num_digits = len(str(num_chunks))
print(num_digits)
ids = [fname + "_" + str(i).zfill(num_digits) for i in range(num_chunks)]
print(ids)
metadata = [{'filename': fname, 
             'bg_name': bg_name, 
             'bgg_id': bgg_id, 
             'rules_forum_id': rules_forum_id,
             'general_forum_id': general_forum_id,
             } for _ in range(num_chunks)]
print(metadata)
```

```python
collection.upsert(
    documents=chunks,
    # embeddings=embeddings.tolist(),
    metadatas=metadata,
    ids=ids
)
```
